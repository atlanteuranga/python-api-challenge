### Store Part I results into DataFrame

```python
csvpath = "../WeatherPy/city_data.csv"
city_df = pd.read_csv(csvpath)
city_df
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
gmaps.configure(api_key=g_key)
locations = city_df[["Latitude", "Longitude"]]
weight = city_df['Humidity']/10
```

```python
figure_layout = {
    'width': '500px',
    'height': '300px',
    'border': '1px solid black',
    'padding': '1px',
    'margin': '0 auto 0 auto'}


# Create heat layer
heat_layer = gmaps.heatmap_layer(locations=locations, weights=weight, 
                                 dissipating=True, max_intensity=10,point_radius=5)

fig=gmaps.figure(layout=figure_layout, zoom_level=1.2, center = (20,0))
# Add layer
fig.add_layer(heat_layer)

# Display figure
fig
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
hotel_df = pd.DataFrame({})
hotel_df['Hotel Name'] = []
hotel_df
```

```python
from pprint import pprint

hotel_names = []
hotel_cities = []
hotel_country = []
hotel_lat = []
hotel_lng = []

for row in desired_df.index:
    target_coordinates = f"{desired_df.loc[row, 'Latitude']},{desired_df.loc[row, 'Longitude']}"
    target_search = "Hotel"
    rankby = "distance"
    target_radius = 5000
    target_type = "hotel"

    # set up a parameters dictionary
    params = {
        "location": target_coordinates,
        "keyword": target_search,
        "radius": target_radius,
        "type": target_type,
        "key": g_key
    }
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    response = requests.get(base_url, params=params).json()
    try:
        hotel_names.append(response['results'][0]['name'])
        hotel_lat.append(response['results'][0]['geometry']['location']['lat'])
        hotel_lng.append(response['results'][0]['geometry']['location']['lng'])
        hotel_cities.append(desired_df.loc[row, 'City'])
    except:
        print(f"No hotels found near {desired_df.loc[row,'City']}")
    
coordinates = []

for i in range(len(hotel_lat)):
    coordinates.append((hotel_lat[i], hotel_lng[i]))
coordinates
```
